```python
import pandas as pd

df = pd.read_csv("LifeTrack White Stork SW Germany_2013-2019-gps.csv")
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.groupby(df['timestamp'].dt.date).first().reset_index(drop=True)
df
```

```python
df
```

```python
df = df[["timestamp", "location-long", "location-lat", "individual-local-identifier"]]
```

```python
df['individual-local-identifier'] = df['individual-local-identifier'].apply(lambda x: x.split(" ")[0])
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['timestamp'] = df['timestamp'].dt.strftime('%Y-%m-%d %H:%M')
```

```python
df.rename(columns={'location-long': 'location_longitude',
                   'location-lat': 'location_latitude',
                  'individual-local-identifier': 'stork_name'}, inplace=True)
```

```python
columns_list = ['timestamp', 'location_longitude', 'location_latitude', 'stork_name']
df = df[columns_list]
```

```python
Mirabell = df[df['stork_name'] == 'Mirabell']
Mirabell

```

```python
mirabell_2016 = Mirabell[Mirabell['timestamp'].str.contains('2016')]
mirabell_2016 
```

```python
mirabell_2016 = mirabell_2016.reset_index(drop=True)
mirabell_2016
```
